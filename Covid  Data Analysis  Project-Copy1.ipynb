```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace=True, axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d' )
```

```python
covid_df.head()
```

```python
covid_df.tail()
```

```python
covid_df['Active_cases'] = covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
```

```python
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values=["Confirmed", "Cured", "Deaths"], index="State/UnionTerritory", aggfunc=max)
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by="Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
top_10_Active_cases = covid_df.groupby('State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by='Active_cases', ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(20,10))
plt.title("Top 10 Active Cases based on statewise", size=25)
ax = sns.barplot(data=top_10_Active_cases.iloc[:10], y = 'Active_cases', x = 'State/UnionTerritory')   
plt.ylabel("Active_cases")
plt.xlabel("State/UnionTerritory")
plt.show()
```

```python
top_10_Death_cases = covid_df.groupby('State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by='Deaths', ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(20,10))
plt.title("Top 10 Death Cases based on statewise", size=25)
ax = sns.barplot(data=top_10_Death_cases.iloc[:12], y = 'Deaths', x = 'State/UnionTerritory')   
plt.ylabel("Death_cases")
plt.xlabel("State/UnionTerritory")
plt.show()
```

```python
vaccine = vaccine_df[vaccine_df.State!='India']
```

```python
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated" : "Total"}, inplace = True)
```

```python
vaccine
```

```python
most_vac = vaccine.groupby('State')['Total'].sum()
```

```python
most_vac = most_vac.to_frame('Total')
```

```python
most_vac
```

```python
most_vac1 = most_vac.sort_values('Total', ascending = False).head()
least_vac1= most_vac.sort_values('Total', ascending = False).tail()
```

```python
fig = plt.figure(figsize=(15, 8))
plt.title("Top 5 most vaccinated states", size=40)
ax = sns.barplot(data = most_vac1,x = most_vac1.index,y = most_vac1.Total, linewidth=2, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

```python
fig = plt.figure(figsize=(14, 8))
plt.title("Least 5 vaccinated states", size=30)
ax = sns.barplot(data = least_vac1,x = least_vac1.index,y = least_vac1.Total, linewidth=2, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
