```python
df[df['Event name'] == 'Yankee Springs 50 km Winter Challenge (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] ==  2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Athlete age'] = 2020 - df2['Athlete year of birth']
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
df2 = df2.drop(columns = ['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2[df2['Athlete age'].isnull()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.isnull().sum()
```

```python
df2.reset_index(drop=True).head()
```

```python
df2.dtypes
```

```python
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2 = df2.rename(columns = { 'Year of event' : 'year',
                              'Event dates':'race_day',
                              'Event name'    : 'race_name',               
                              'Event distance/length' : 'race_number_of_finishers',      
                              'Event number of finishers' : 'athlete_performance',    
                              'Athlete performance' :  'athlete_performance',        
                              'Athlete gender' : 'athlete_gender',               
                              'Athlete average speed' : 'athlete_average_speed',       
                              'Athlete ID' : 'athlete_id',                   
                              'Athlete age' : 'athlete_age'               
                             })
```

```python
df2.head()
```

```python
df3 = df2[['race_day', 'race_name', 'race_number_of_finishers', 'athlete_performance', 'athlete_performance', 'athlete_gender', 'athlete_average_speed', 'athlete_id', 'athlete_age']]
```

```python
sns.histplot(df3['race_length'])
```

```python
sns.histplot(df3, x = 'race_length' , hue = 'athlete_gender')
```

```python
sns.violinplot(data = df3, x = 'race_length', y = 'athlete_average_speed', hue='athlete_gender', split=True, inner = 'quart', linewidth =2)
```

```python
sns.lmplot(data=df3, x = 'athlete_age', y = 'athlete_average_speed', hue='athlete_gender')
```

```python
## what age groups are the best in the 50m Race (20 + races min)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count>19').head(15)
```
